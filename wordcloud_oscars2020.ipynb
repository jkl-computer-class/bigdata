```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
text = ""
with open('oscars2020.txt', encoding='utf-8') as f:
    text = ''.join(f.readlines())
```

```python
wc = WordCloud()
```

```python
wc.generate(text)
```

```python
wc.to_file('output.png')
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(max_font_size=40).generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
custom_mask = np.array(Image.open("trophy.png"))
wc = WordCloud(background_color="black", mask=custom_mask)
wc.generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(background_color="white", mask=custom_mask, contour_width=3, contour_color='steelblue')
```

```python
custom_mask = np.array(Image.open("trophy.png"))
wc = WordCloud(background_color="white", mask=custom_mask, contour_width=2, contour_color='gold')
wc.generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
custom_mask = np.array(Image.open("trophy.png"))
wc = WordCloud(background_color="white", mask=custom_mask)
wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func=image_colors)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import re
from collections import Counter
```

```python
words = re.findall(r'\w+', open('oscars2020.txt', encoding='utf-8').read().lower())
words_counter = Counter(words)
```

```python
wc = WordCloud(background_color="white", mask=custom_mask)
wc.generate_from_frequencies(words_counter)
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
